```python
import numpy as np
import pandas as pd
```

```python
properties = pd.read_csv('domain_properties.csv')
suburbs = pd.read_csv('Sydney-Suburbs-Reviews.csv')
```

```python
properties.head()
```

```python
properties.info()
```

```python
# just taking the absolute value of prices, just in case
properties['price'] = np.abs(properties['price'])
```

```python
properties.info()
```

```python
properties = properties.drop(axis = 1, columns = ['date_sold', 'suburb_sqkm', 'suburb_lat','suburb_lng', 'suburb_elevation'])
```

```python
properties.head()
```

```python
# different types of homes
properties['type'].value_counts()
```

```python
# let's drop all rows that are not houses
properties = properties[properties['type'] == 'House']
properties.drop(axis = 1, columns = ['type'], inplace = True)
properties.info()
```

```python
properties.head()
```

```python
properties['suburb'].value_counts()
```

```python
# let's drop all suburbs with less than 10 properties
properties = properties[properties['suburb'].map(properties['suburb'].value_counts()) >= 10]
properties['suburb'].value_counts()
```

```python
# ist of all suburbs with more than 10 properties
suburb_list = properties['suburb'].unique()
list(suburb_list)
```

```python
# suburb, it's population, and median income
suburb_info = properties[['suburb','suburb_population', 'suburb_median_income']]

#removing duplicates
suburb_info = suburb_info.drop_duplicates()
suburb_info.info()

suburb_info.to_csv('suburb_info.csv', index = False)
```

```python
properties.info()
```

```python
# features and target
X = properties.drop(axis = 1, columns = ['price','suburb'])
y = properties['price']
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=10)
```

```python
# # onehot encoding the suburb
# from sklearn.compose import make_column_transformer
# from sklearn.preprocessing import OneHotEncoder
# ct = make_column_transformer((OneHotEncoder(), ['suburb']), remainder = 'drop')

# #making a pipeline containing this onehot encoder and xgboost regressor
# from sklearn.pipeline import make_pipeline
# from xgboost import XGBRegressor
# model = make_pipeline(ct, XGBRegressor())
# model.fit(X_train, y_train)
```

```python
# define and fit xgboost regressor
from xgboost import XGBRegressor
model = XGBRegressor(n_estimators = 160, random_state = 10, max_depth = 3, learning_rate = 0.1)
model.fit(X_train, y_train)
```

```python
# evaluate the model with r2, mae, rmse, and mape, against both the train and test sets
from sklearn.metrics import r2_score, mean_absolute_error, root_mean_squared_error, mean_absolute_percentage_error

print(f"Train r2 score: {r2_score(y_train, model.predict(X_train))}")
print(f"Test r2 score: {r2_score(y_test, model.predict(X_test))}")
print(f"Train mae: {mean_absolute_error(y_train, model.predict(X_train))}")
print(f"Test mae: {mean_absolute_error(y_test, model.predict(X_test))}")
print(f"Train rmse: {root_mean_squared_error(y_train, model.predict(X_train))}")
print(f"Test rmse: {root_mean_squared_error(y_test, model.predict(X_test))}")
print(f"Train mape: {mean_absolute_percentage_error(y_train, model.predict(X_train))}")
print(f"Test mape: {mean_absolute_percentage_error(y_test, model.predict(X_test))}")
```

```python
# pickling out the model
import pickle
pickle.dump(model, open('model.pkl', 'wb'))
```
